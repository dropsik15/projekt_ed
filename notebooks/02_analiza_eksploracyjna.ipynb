```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

df = pd.read_csv("../data/processed/mieszkania_clean.csv")
df["year_month"] = pd.to_datetime(df["year_month"])
```

```python
price_trend = df.groupby("year_month")["price"].median().reset_index()

plt.figure(figsize=(10,5))
sns.lineplot(data=price_trend, x="year_month", y="price", marker="o")
plt.title("Mediana ceny mieszkania w czasie")
plt.ylabel("Cena [zł]")
plt.xlabel("Miesiąc")
plt.xticks(rotation=45)

plt.gca().yaxis.set_major_formatter(
    mtick.FuncFormatter(lambda x, _: f"{int(x):,} zł".replace(",", " "))
)

plt.tight_layout()
plt.savefig("../figures/trend_ceny_mieszkania.png", dpi=300)
plt.show()
```

```python
ppm_trend = df.groupby("year_month")["price_per_m2"].median().reset_index()

plt.figure(figsize=(10,5))
sns.lineplot(data=ppm_trend, x="year_month", y="price_per_m2", marker="o")
plt.title("Mediana ceny za m² w czasie")
plt.ylabel("Cena za m² [zł]")
plt.xlabel("Miesiąc")
plt.xticks(rotation=45)

plt.gca().yaxis.set_major_formatter(
    mtick.FuncFormatter(lambda x, _: f"{int(x):,} zł/m²".replace(",", " "))
)

plt.tight_layout()
plt.savefig("../figures/trend_ceny_m2.png", dpi=300)
plt.show() 
```
